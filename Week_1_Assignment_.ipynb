# **Descriptive Stats**

```python
math_scores = df['Math']

mean = math_scores.mean()
median = math_scores.median()
mode = math_scores.mode()[0]
std_dev = math_scores.std()

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std_dev)
```

```python
Science_scores = df['Science']

mean = Science_scores.mean()
median = Science_scores.median()
mode = Science_scores.mode()[0]
std_dev = Science_scores.std()

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std_dev)
```

```python
English_scores = df['English']

mean = English_scores.mean()
median = English_scores.median()
mode = English_scores.mode()[0]
std_dev = English_scores.std()

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std_dev)
```

```python
History_scores = df['History']

mean = History_scores.mean()
median = History_scores.median()
mode = History_scores.mode()[0]
std_dev = History_scores.std()

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std_dev)
```

```python
Computer_Science_scores = df['Computer_Science']

mean = Computer_Science_scores.mean()
median = Computer_Science_scores.median()
mode = Computer_Science_scores.mode()[0]
std_dev = Computer_Science_scores.std()

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", std_dev)
```

# **Visualization**

```python
subject_columns = ['Math', 'Science', 'English', 'History', 'Computer_Science']

for subject_col in subject_columns:
    plt.figure()
    sns.histplot(df[subject_col], kde=True)
    plt.title(f"Distribution of {subject_col} Scores")
    plt.xlabel("Marks")
    plt.ylabel("Frequency")
    plt.show()
```

1. The average (mean) score is around 69.56, indicating overall student performance is moderate to good.

3. The standard deviation is 20.11, which shows high variation in scores.
